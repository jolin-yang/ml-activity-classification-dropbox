```python
features = pd.read_csv("/content/drive/MyDrive/UCI Dataset/features.txt", sep=r"\s+", header = None)
feature_names = features[1].tolist()
print(feature_names)
```

```python
X_train = pd.read_csv("/content/drive/MyDrive/UCI Dataset/X_train.txt", sep=r"\s+", header=None)
X_train.columns = feature_names

print("Shape of X_train:", X_train.shape)

X_train.head()
```

```python
activity_df = pd.read_csv("/content/drive/MyDrive/UCI Dataset/activity_labels.txt", sep=r"\s+", header = None)
print(activity_df[0].tolist())
print(activity_df[1].tolist())
```

```python
y_train = pd.read_csv("/content/drive/MyDrive/UCI Dataset/y_train.txt", sep=r"\s+", header=None, names = ["Activity Label"])
y_train['Activity'] = y_train['Activity Label'].map({
    1: 'WALKING',
    2: 'WALKING_UPSTAIRS',
    3: 'WALKING_DOWNSTAIRS',
    4: 'SITTING',
    5: 'STANDING',
    6: 'LAYING'
})

print("Shape of y_train:", y_train.shape)

y_train.head(10)
```

```python
X_test = pd.read_csv("/content/drive/MyDrive/UCI Dataset/X_test.txt", sep=r"\s+", header=None)
X_test.columns = feature_names

print("Shape of X_test:", X_test.shape)

X_test.head()
```

```python
y_test = pd.read_csv("/content/drive/MyDrive/UCI Dataset/y_test.txt", sep=r"\s+", header=None, names = ["Activity Label"])
y_test['Activity'] = y_test['Activity Label'].map({
    1: 'WALKING',
    2: 'WALKING_UPSTAIRS',
    3: 'WALKING_DOWNSTAIRS',
    4: 'SITTING',
    5: 'STANDING',
    6: 'LAYING'
})

print("Shape of y_test:", y_test.shape)

y_test.head(10)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
lr_model = LogisticRegression(max_iter = 500)
lr_model.fit(X_train, y_train['Activity Label'])
y_predictions = lr_model.predict(X_test)
accuracy = accuracy_score(y_test['Activity Label'], y_predictions)
print("Accuracy of Logistic Regression model before performing feature engineering and removing highly correlated features:", accuracy)
```

```python
feature_subset = X_train.iloc[:, 344:459]
corr_subset = feature_subset.corr()
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(corr_subset)
plt.title("Correlation heatmap of Features 345-460")
plt.show()
```

```python
corr_matrix = feature_subset.corr().abs()
upper_tri = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
highly_correlated_pairs = upper_tri.stack().loc[lambda x: x >= 0.9]

print("Highly correlated feature pairs where correlation is >= 0.9:")
print(highly_correlated_pairs)
```

```python
# drop highly correlated features
dropped_features = list(highly_correlated_pairs.index.get_level_values(1).unique())
reduced_features_subset = feature_subset.drop(columns=dropped_features)
```

```python
print(reduced_features_subset.shape)
print("We have gone from 120 features to 43 after removing highly correlated pairs in the subset of features 345-460.")
```

```python
print("Kept features in the subset:")
for col in reduced_features_subset.columns:
    print(col)
```

```python
X_train_reduced = X_train.drop(columns=dropped_features)
X_test_reduced = X_test.drop(columns=dropped_features)
```

```python
lr_model = LogisticRegression(max_iter = 500)
lr_model.fit(X_train_reduced, y_train['Activity Label'])
y_predictions = lr_model.predict(X_test_reduced)
accuracy = accuracy_score(y_test['Activity Label'], y_predictions)
print("Accuracy of Logistic Regression model after performing feature engineering and removing highly correlated features in subset:", accuracy)
```

**From these findings:**

Accuracy of Logistic Regression model before performing feature engineering and removing highly correlated features: 0.9602986087546658

Accuracy of Logistic Regression model after performing feature engineering and removing highly correlated features: 0.9609772650152698

There is a slight, but very miniscule improvement in the accuracy score (~.007) after performing feature engineering on the subset of features and removing highly-correlated features.
